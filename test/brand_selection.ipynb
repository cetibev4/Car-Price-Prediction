# The dataset has been cleaned by keeping top Manufacturer and brands here

```python
import pandas as pd

df = pd.read_excel('../car_price_prediction.xlsx')

numeric_features = ['Levy', 'Prod. year', 'Engine volume', 'Mileage', 'Cylinders', 'Airbags']
categorical_features = ['Manufacturer', 'Category', 'Fuel type', 'Gear box type', 'Drive wheels', 'Color']
```

```python
import pandas as pd

pd.set_option('display.max_rows', None)

print(df['Manufacturer'].value_counts())
```

```python
top_brands = [
'HYUNDAI','TOYOTA','MERCEDES-BENZ','FORD','CHEVROLET','BMW','LEXUS',
'HONDA','NISSAN','VOLKSWAGEN','SSANGYONG','KIA','OPEL','MITSUBISHI',
'SUBARU','AUDI','MAZDA','JEEP','DAEWOO','DODGE','FIAT','SUZUKI',
'PORSCHE','LAND ROVER'
] #brands we want to keep
df = df[df['Manufacturer'].isin(top_brands)] # filter dataframe

df['Manufacturer'].value_counts()
```

```python
pd.set_option('display.max_rows', None)

print(df['Model'].value_counts())
```

```python
top_models =  ["Prius", "Sonata", "Camry", "Elantra", "E 350", "Santa FE", "FIT", "H1",
              "Tucson", "X5", "Aqua", "Cruze", "Fusion", "Optima", "GX 460", "Transit",
              "Highlander", "ML 350", "Jetta", "Actyon", "Civic", "REXTON", "RX 450", "RAV 4"]

df = df[df['Model'].isin(top_models)] # filter dataframe

df['Model'].value_counts()  

df.to_excel('model_counts.xlsx', index=False)
```
